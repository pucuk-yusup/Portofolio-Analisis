```python
# Import pustaka
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

```python
# Cek versi library untuk memastikan semuanya terinstall
print("NumPy Version:", np.__version__)
print("Seaborn Version:", sns.__version__)
print("Matplotlib Version:", matplotlib.__version__)
```

```python
# Buat array dengan NumPy
arr = np.random.randint(1, 100, (10,))  # 10 angka random
print("Array Random NumPy:", arr)
```

```python
# Visualisasi Data dengan Seaborn
df = sns.load_dataset("penguins")  # Load dataset contoh
sns.histplot(df['flipper_length_mm'].dropna(), kde=True, bins=20)
plt.title("Distribusi Panjang Flipper Penguin")
plt.show()
```

```python
# Plot Sederhana dengan Matplotlib
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="Sin(x)")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.title("Grafik Sukses dengan Matplotlib")
plt.show()
```
