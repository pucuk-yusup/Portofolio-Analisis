```python
import importlib

# Daftar pustaka yang sering digunakan oleh Data Analyst
libraries = [
    "numpy", "pandas", "matplotlib", "seaborn", "scipy",
    "statsmodels", "scikit-learn", "plotly", "openpyxl",
    "xlrd", "requests", "beautifulsoup4", "sqlalchemy"
]
```

```python
# Cek apakah pustaka sudah terinstal
missing_libs = []
for lib in libraries:
    if importlib.util.find_spec(lib) is None:
        missing_libs.append(lib)
```

```python
# Output hasil pengecekan
if not missing_libs:
    print("✅ Semua pustaka Data Analyst sudah terinstal!")
else:
    print("⚠️ Pustaka berikut belum terinstal:")
    for lib in missing_libs:
        print(f"- {lib}")
    print("\nGunakan perintah berikut untuk menginstal yang belum ada:")
    print(f"!pip install {' '.join(missing_libs)}")  # Perintah khusus untuk Jupyter Notebook
```
